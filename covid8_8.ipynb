```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/datasets/covid-19/main/data/countries-aggregated.csv"
data = pd.read_csv(url, parse_dates=['Date'])
```

```python
print(data.head())
```

```python
country_data = data[data['Country'] == 'India']
```

```python
print(country_data.head())
```

```python
plt.figure(figsize = (10,6))
plt.plot(country_data['Date'], country_data['Confirmed'], label = 'Confirmed Cases')
plt.title('COVID-19 Cases and Deaths in India')
plt.xlabel('Date')
plt.ylabel('Number of cases')
```

```python
plt.plot(country_data['Date'], country_data['Deaths'], label = 'Deaths')
plt.title('COVID-19 Cases and Deaths in India')
plt.xlabel('Date')
plt.ylabel('Number of deaths')
```

## Seasonal and trend analysis

```python
country_data = data[data['Country'] == 'India'].copy()
country_data['Date'] = pd.to_datetime(country_data['Date'])

country_data.set_index('Date', inplace=True)
monthly_data = country_data['Confirmed'].resample('M').sum()
death_data = country_data['Deaths'].resample('M').sum()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decomposition = seasonal_decompose(monthly_data, model = 'additive')
fig = decomposition.plot()
fig.set_size_inches(8,8)
plt.show()
```

```python
decomposition = seasonal_decompose(monthly_data, model = 'multiplicative')
fig = decomposition.plot()
fig.set_size_inches(8,8)
plt.show()
```

```python
decomposition_deaths = seasonal_decompose(death_data, model = 'additive')
fig1 = decomposition_deaths.plot()
fig1.set_size_inches(8,8)
plt.show()
```
